- Test your script by assigning values with different data types to the variable.

```python
user_variable = input("What would you like initialize? ")

if(user_variable.isdigit()):
    print(f"This is an integer!")
elif(user_variable.find(".") >= 0 and user_variable.count(".") == 1 and user_variable[:(user_variable.find("."))].isdigit() and user_variable[(user_variable.find(".")+1):].isdigit()):
    print(f"This is a float!")
else:
    print(f"This is a string!")
```

# Section 3

Gate admission categories for the Oklahoma State Fair are listed in this table:

- Age Range: Admission ($)

- Ages 5 and under:	FREE!

- Ages 6-11:	6

- Ages 12+:	12

Write a script that uses the age of the fair visitor as input and prints the admission as output to the screen. Use the input() function, logical operators, and a chained conditional to implement this script. Double-check the ranges of the three different categories.

```python
def ticket_price(age):
    """Determines the price of the ticket by the age person attending the fair."""
    if age < 0:
        return -1
    elif age <= 5:
        return 0
    elif age <= 11:
        return 6
    elif age >= 12:
        return 12
    else:
        return -1
    
    
another_ticket = "Y"
total = 0
number_of_tickets = 0

while another_ticket[0] == "Y":
    ticket_age = input("On their last birthday, how old is the visitor? ")
    if ticket_age.isdigit():
        price = ticket_price(int(ticket_age))
        
        if price != -1:
            number_of_tickets += 1
            total += price
            print(f"That ticket is ${price}.")
        
            another_ticket = input("Want another ticket Y/N? ").upper()
        else:
            another_ticket = "N"
            break
    else:
        another_ticket = "N"
        break
    
print(f"You want to purchase {number_of_tickets} tickets for a total of ${total}.")
```
